```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
metrics_our = pd.read_csv("history_our.csv")
metrics_main = pd.read_csv("history_main.csv")
```

```python
metrics_our.describe()
```

```python
metrics_main.describe()
```

# Main model metrics

```python
metrics_main[['dice_score', 'val_dice_score']].plot()
```

Both training and validation dice scores increase over time, which indicates that the model is learning and improving its ability to segment correctly. However, the validation score begins to flatten or slightly decrease after some epochs, suggesting possible overfitting (i.e., the model performs well on training data but less so on unseen validation data).

```python
metrics_main[['lr']].plot()
```

The learning rate starts high, and at around epoch 25, it significantly decreases and continues to drop until the end of training. This pattern suggests that a learning rate scheduler is in place, which reduces the learning rate after a certain number of epochs or when performance plateaus, to allow for finer adjustments to the model's parameters.

```python
metrics_main[['loss', 'val_loss']].plot()
```

Both training and validation loss decrease rapidly at the beginning and then continue to slowly decline, showing that the model is learning. The fact that training and validation loss curves are close together indicates that the model is generalizing well and not overfitting (no significant gap between the curves).

# Our model metrics

```python
metrics_our[['dice_score', 'val_dice_score']].plot()
```

The training dice score (blue line) shows a gradual increase, indicating that the model is improving in its ability to segment during training. However, it starts to fluctuate around epoch 15.
The validation dice score (orange line) remains fairly constant after initially increasing, which could indicate the model is not improving significantly on the validation set beyond a certain point. There might be some overfitting here, as the validation score plateaus while the training score fluctuates.

```python
metrics_our[['loss', 'val_loss']].plot()
```

Both training loss and validation loss drop sharply within the first few epochs and continue to decline at a much slower pace, stabilizing after about 5 epochs.
The fact that both loss curves follow a similar trend indicates that the model is learning in a balanced manner without significant overfitting.


```python
metrics_our[['lr']].plot()
```

Initially, the learning rate is constant at a high value, but after epoch 10, it drops and then continues to decrease further after epoch 15.
This behavior suggests the use of a learning rate schedule, where the learning rate is reduced to help the model converge with finer adjustments as training progresses.
